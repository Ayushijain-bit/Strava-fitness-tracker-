```python
import matplotlib.pyplot as plt

test_dates = ['2025-06-01', '2025-06-02', '2025-06-03']
asleep = [350, 400, 390]
in_bed = [420, 450, 430]

plt.figure(figsize=(8,5))
plt.plot(test_dates, asleep, marker='o', label='Minutes Asleep', color='blue')
plt.plot(test_dates, in_bed, marker='s', label='Time in Bed', color='orange')
plt.title('Dummy Sleep Duration Test')
plt.xlabel('Date')
plt.ylabel('Minutes')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**3. Weight and BMI Trend**

```python
weight_df = merged_df[['Date', 'WeightKg', 'BMI']].dropna()
weight_df['Date'] = pd.to_datetime(weight_df['Date'], dayfirst=True)

plt.figure(figsize=(12, 6))
plt.plot(weight_df['Date'], weight_df['WeightKg'], label='Weight (Kg)', marker='o')
plt.plot(weight_df['Date'], weight_df['BMI'], label='BMI', marker='x')
plt.title('Weight and BMI Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**4. Average sleep, calories, and weight**

```python
avg_sleep = merged_df['TotalMinutesAsleep_x'].mean()
avg_calories = merged_df['Calories_x'].mean()
avg_weight = merged_df['WeightKg'].mean()

labels = ['Avg Sleep (min)', 'Avg Calories', 'Avg Weight (kg)']
values = [avg_sleep, avg_calories, avg_weight]
colors = ['skyblue', 'lightgreen', 'salmon']

plt.figure(figsize=(8, 5))
bars = plt.bar(labels, values, color=colors)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 5, f'{yval:.1f}', ha='center', fontsize=12)

plt.title('Average Sleep, Calories Burned, and Weight')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

**5. Average sleep time**

```python
avg_sleep = merged_df['TotalMinutesAsleep_x'].mean()

plt.figure(figsize=(4, 5))
plt.bar(['Average Sleep'], [avg_sleep], color='skyblue')
plt.ylabel('Minutes')
plt.title('Average Sleep Time')
plt.ylim(0, max(avg_sleep + 50, 500))  # set upper limit slightly above value
plt.text(0, avg_sleep + 10, f'{avg_sleep:.1f} min', ha='center', fontsize=12)
plt.tight_layout()
plt.show()
```

**6. Weight Overtime**

```python
weight_df['Date'] = pd.to_datetime(weight_df['Date'], dayfirst=True)

weight_df = weight_df.sort_values('Date')

# Plot
plt.figure(figsize=(10, 5))
plt.plot(weight_df['Date'], weight_df['WeightKg'], marker='o', color='green', linestyle='-')
plt.title('Weight Over Time')
plt.xlabel('Date')
plt.ylabel('Weight (Kg)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

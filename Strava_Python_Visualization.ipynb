**6.Average Weight or BMI**

```python
average_weight = df['WeightKg'].mean()
print(f'The average weight is:{average_weight}')

average_bmi = df['BMI'].mean()
print(f'The average BMI is:{average_bmi}')
```

```python
labels = ['Average Weight (kg)', 'Average BMI']
average_values = [average_weight, average_bmi]

plt.figure(figsize=(8, 5))
plt.bar(labels, average_values, color=['skyblue', 'lightgreen'])
plt.ylabel('Value')
plt.title('Average Weight and BMI')
plt.ylim(0, max(average_values) * 1.2)

for i, v in enumerate(average_values):
    plt.text(i, v + 0.5, str(round(v, 2)), ha='center', va='bottom')

plt.show()
```
